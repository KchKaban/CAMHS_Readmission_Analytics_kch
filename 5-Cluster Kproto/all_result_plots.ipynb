```python
%run -i allpackages.py

class Cluster_Assignment:
    def __init__(self):
        self.original_df_1 = pd.read_csv(
            "ClusterLabel_Full_ICD10_ATC_Dummies_ICD10_ATC_20.csv"
        )

        self.original_df_Full_ICD10_ATC = pd.read_csv(
            "new-data/Full_ICD10_ATC.csv"
        )

    def load_data_with_top50_diag_med(self):
        original_df_1merged_df = self.original_df_Full_ICD10_ATC[
            [
                "pasient",
                "episode_id",
                "episode_start_date",
                "episode_end_date",
                "gender",
                "age",
                "diagnosis",
                "actual_med_Full_ATC",
                "Length_of_Episode",
                "Count_visit",
                "var_no_dates_permonth",
                "num_diagnoses",
                "num_medications",
                "Inpatient_day_ratio",
                "Outpatient_ratio",
                "Therapy_ratio",
                "TreatmentPlanning_ratio",
                "Advisory_ratio",
                "age_group",
                "Examination_ratio",
                "Inpatient_ratio",
                "Inpatient_daynight_ratio",
                "tillnextepisode",
            ]
        ].merge(
            self.original_df_1[["episode_id", "cluster", "cluster_distances"]],
            on="episode_id",
            how="inner",
        )

        # Rename the column var_no_dates_permonth with the Intensity_per_calendar_month
        original_df_1merged_df = original_df_1merged_df.rename(
            columns={"var_no_dates_permonth": "SD_CareEvent_PerMonth"}
        )
        le = LabelEncoder()
        original_df_1merged_df["tillnextepisode_bins"] = le.fit_transform(
            pd.cut(
                original_df_1merged_df["tillnextepisode"],
                bins=[float("-inf"), 0, 182, 730, float("inf")],
                labels=[
                    "not-re-admitted",
                    "re-admitted in 0-182 days",
                    "re-admitted in 182-730 days",
                    "re-admitted in more than 730 days",
                ],
            )
        )

        original_df_1merged_df = original_df_1merged_df.rename(
            columns={"age": "Age",
                     "num_diagnoses": "Num_diagnoses",
                        "num_medications": "Num_medications",
                        "tillnextepisode_bins": "Tillnextepisode_bins"
                     }
        )

        return original_df_1merged_df

    def calculate_max_min(self, original_df_1merged_df):

        # Calculate the minimum and maximum value of the columns
        min_max = original_df_1merged_df[
            [
                "Age",
                "Length_of_Episode",
                "Count_visit",
                "SD_CareEvent_PerMonth",
                "Num_diagnoses",
                "Num_medications",
                "Inpatient_day_ratio",
                "Outpatient_ratio",
                "Therapy_ratio",
                "TreatmentPlanning_ratio",
                "Advisory_ratio",
                "age_group",
                "Examination_ratio",
                "Inpatient_ratio",
                "Inpatient_daynight_ratio",
                "Tillnextepisode_bins",
            ]
        ].agg(["min", "max"])

        # Display the result
        display(min_max)


Cluster_Assignment_inst = Cluster_Assignment()
original_df_1merged_df = Cluster_Assignment_inst.load_data_with_top50_diag_med()
original_df_1merged_df["Age"] = (
    original_df_1merged_df["Age"] - original_df_1merged_df["Age"].min()
) / (original_df_1merged_df["Age"].max() - original_df_1merged_df["Age"].min())
original_df_1merged_df["Inpatient_day_ratio"] = (
    original_df_1merged_df["Inpatient_day_ratio"]
    - original_df_1merged_df["Inpatient_day_ratio"].min()
) / (
    original_df_1merged_df["Inpatient_day_ratio"].max()
    - original_df_1merged_df["Inpatient_day_ratio"].min()
)
original_df_1merged_df["Outpatient_ratio"] = (
    original_df_1merged_df["Outpatient_ratio"]
    - original_df_1merged_df["Outpatient_ratio"].min()
) / (
    original_df_1merged_df["Outpatient_ratio"].max()
    - original_df_1merged_df["Outpatient_ratio"].min()
)
original_df_1merged_df["Therapy_ratio"] = (
    original_df_1merged_df["Therapy_ratio"]
    - original_df_1merged_df["Therapy_ratio"].min()
) / (
    original_df_1merged_df["Therapy_ratio"].max()
    - original_df_1merged_df["Therapy_ratio"].min()
)
original_df_1merged_df["TreatmentPlanning_ratio"] = (
    original_df_1merged_df["TreatmentPlanning_ratio"]
    - original_df_1merged_df["TreatmentPlanning_ratio"].min()
) / (
    original_df_1merged_df["TreatmentPlanning_ratio"].max()
    - original_df_1merged_df["TreatmentPlanning_ratio"].min()
)
original_df_1merged_df["Advisory_ratio"] = (
    original_df_1merged_df["Advisory_ratio"]
    - original_df_1merged_df["Advisory_ratio"].min()
) / (
    original_df_1merged_df["Advisory_ratio"].max()
    - original_df_1merged_df["Advisory_ratio"].min()
)
original_df_1merged_df["Length_of_Episode"] = (
    original_df_1merged_df["Length_of_Episode"]
    - original_df_1merged_df["Length_of_Episode"].min()
) / (
    original_df_1merged_df["Length_of_Episode"].max()
    - original_df_1merged_df["Length_of_Episode"].min()
)
original_df_1merged_df["Count_visit"] = (
    original_df_1merged_df["Count_visit"] - original_df_1merged_df["Count_visit"].min()
) / (
    original_df_1merged_df["Count_visit"].max()
    - original_df_1merged_df["Count_visit"].min()
)
original_df_1merged_df["SD_CareEvent_PerMonth"] = (
    original_df_1merged_df["SD_CareEvent_PerMonth"]
    - original_df_1merged_df["SD_CareEvent_PerMonth"].min()
) / (
    original_df_1merged_df["SD_CareEvent_PerMonth"].max()
    - original_df_1merged_df["SD_CareEvent_PerMonth"].min()
)
original_df_1merged_df["Num_diagnoses"] = (
    original_df_1merged_df["Num_diagnoses"]
    - original_df_1merged_df["Num_diagnoses"].min()
) / (
    original_df_1merged_df["Num_diagnoses"].max()
    - original_df_1merged_df["Num_diagnoses"].min()
)
original_df_1merged_df["Num_medications"] = (
    original_df_1merged_df["Num_medications"]
    - original_df_1merged_df["Num_medications"].min()
) / (
    original_df_1merged_df["Num_medications"].max()
    - original_df_1merged_df["Num_medications"].min()
)
original_df_1merged_df["Tillnextepisode_bins"] = (
    original_df_1merged_df["Tillnextepisode_bins"]
    - original_df_1merged_df["Tillnextepisode_bins"].min()
) / (
    original_df_1merged_df["Tillnextepisode_bins"].max()
    - original_df_1merged_df["Tillnextepisode_bins"].min()
)


cluster0 = original_df_1merged_df[original_df_1merged_df["cluster"] == 0]
cluster1 = original_df_1merged_df[original_df_1merged_df["cluster"] == 1]
cluster2 = original_df_1merged_df[original_df_1merged_df["cluster"] == 2]
cluster3 = original_df_1merged_df[original_df_1merged_df["cluster"] == 3]
cluster4 = original_df_1merged_df[original_df_1merged_df["cluster"] == 4]

print("Cluster 0")
Cluster_Assignment_inst.calculate_max_min(cluster0)
print("Cluster 1")
Cluster_Assignment_inst.calculate_max_min(cluster1)
print("Cluster 2")
Cluster_Assignment_inst.calculate_max_min(cluster2)
print("Cluster 3")
Cluster_Assignment_inst.calculate_max_min(cluster3)
print("Cluster 4")
Cluster_Assignment_inst.calculate_max_min(cluster4)

def plot_min_max_all(clusters):
    columns = clusters[0].columns
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_title("Model with 5 clusters: For readmitted and not-readmitted episodes")
    ax.set_xticks(range(len(columns)))
    ax.set_xticklabels(columns, rotation=90)
    colors = ["red", "blue", "green", "orange", "purple"]
    offset = 0.15
    for i, cluster in enumerate(clusters):
        min_values = cluster.min()
        max_values = cluster.max()
        ax.scatter(
            [j + i * offset for j in range(len(columns))],
            min_values,
            color=colors[i],
            marker="v",
        )
        ax.scatter(
            [j + i * offset for j in range(len(columns))],
            max_values,
            color=colors[i],
            marker="^",
        )
        for j in range(len(columns)):
            ax.plot(
                [j + i * offset, j + i * offset],
                [min_values[j], max_values[j]],
                color=colors[i],
                linewidth=1.8,
            )
    handles = [plt.Line2D([0], [0], color=c, lw=4) for c in colors]
    labels = [f"Cluster {i}" for i in range(len(clusters))]
    ax.legend(handles, labels, bbox_to_anchor=(1.04, 1), loc="upper left")
    fig.tight_layout()
    plt.xlabel("Features", fontsize=12)
    plt.ylabel("Normalized Values", fontsize=12)
    plt.savefig("Figure I(a).png", bbox_inches="tight", dpi=300)
    plt.show()


cl0 = cluster0[
    [
        "Age",
        "Inpatient_day_ratio",
        "Outpatient_ratio",
        "Therapy_ratio",
        "TreatmentPlanning_ratio",
        "Advisory_ratio",
        "Length_of_Episode",
        "Count_visit",
        "SD_CareEvent_PerMonth",
        "Num_diagnoses",
        "Num_medications",
        "Tillnextepisode_bins",
    ]
]
cl1 = cluster1[
    [
        "Age",
        "Inpatient_day_ratio",
        "Outpatient_ratio",
        "Therapy_ratio",
        "TreatmentPlanning_ratio",
        "Advisory_ratio",
        "Length_of_Episode",
        "Count_visit",
        "SD_CareEvent_PerMonth",
        "Num_diagnoses",
        "Num_medications",
        "Tillnextepisode_bins",
    ]
]
cl2 = cluster2[
    [
        "Age",
        "Inpatient_day_ratio",
        "Outpatient_ratio",
        "Therapy_ratio",
        "TreatmentPlanning_ratio",
        "Advisory_ratio",
        "Length_of_Episode",
        "Count_visit",
        "SD_CareEvent_PerMonth",
        "Num_diagnoses",
        "Num_medications",
        "Tillnextepisode_bins",
    ]
]
cl3 = cluster3[
    [
        "Age",
        "Inpatient_day_ratio",
        "Outpatient_ratio",
        "Therapy_ratio",
        "TreatmentPlanning_ratio",
        "Advisory_ratio",
        "Length_of_Episode",
        "Count_visit",
        "SD_CareEvent_PerMonth",
        "Num_diagnoses",
        "Num_medications",
        "Tillnextepisode_bins",
    ]
]
cl4 = cluster4[
    [
        "Age",
        "Inpatient_day_ratio",
        "Outpatient_ratio",
        "Therapy_ratio",
        "TreatmentPlanning_ratio",
        "Advisory_ratio",
        "Length_of_Episode",
        "Count_visit",
        "SD_CareEvent_PerMonth",
        "Num_diagnoses",
        "Num_medications",
        "Tillnextepisode_bins",
    ]
]


plot_min_max_all([cl0, cl1, cl2, cl3, cl4])
```
